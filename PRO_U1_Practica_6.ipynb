```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
import math
```

```python
carro = pd.read_csv("autos2.csv")
```

```python
carro.info()
```

```python
media = carro.price.mean()
mediana = carro.price.median()
moda = carro.price.mode()
minimo = carro.price.min()
maximo = carro.price.max()
desviacion = carro.price.std()
df = pd.DataFrame(carro.price)
print (df.describe)
print ('Media de los precios de los coches',(media))
print ('Mediana de los precios de los coches',(mediana))
print ('Moda de los precios de los coches',(moda))
print ('Valor minimo de los precios de los coches',(minimo))
print ('Valor maximo de los precios de los coches',(maximo))
print ('Desviacion estandar de los precios de los coches',(desviacion))
```

```python
cuartiles = df['price'].quantile([0.25, 0.5, 0.75])
print(cuartiles)
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(10,5))
plt.hist(carro.price, bins = 10, alpha = 0.2, color = 'b', edgecolor = 'black')
plt.axvline(media, color = 'r', linestyle = 'dashed', linewidth = 2)
plt.axvline(mediana, color = 'g', linestyle = 'dashed', linewidth = 2)
plt.axvline(moda.iloc[0], color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(minimo, color = 'm', linestyle = 'dashed', linewidth = 2)
plt.axvline(maximo, color = 'c', linestyle = 'dashed', linewidth = 2)
plt.axvline(desviacion, color = 'k', linestyle = 'dashed', linewidth = 2)
plt.title('Histograma de los precios de los coches')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.legend(['Media', 'Mediana', 'Moda', 'Minimo', 'Maximo', 'Desviacion'])
plt.show()
```

```python
n = len(carro.price)
K = int(1 + math.log2(n))
print("Valor de n = ", n , " Número de intervalos =", K)
```

```python
tabla_frecuencia = pd.cut(carro.price, bins=K, include_lowest=True).value_counts().reset_index()
print(tabla_frecuencia)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=carro)
plt.title('Diagrama de Caja y Bigotes: Precio vs. Tipo de Tracción de Ruedas')
plt.xlabel('Tipo de Tracción de Ruedas')
plt.ylabel('Precio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='engine-location', y='price', data=carro)
plt.title('Diagrama de Caja y Bigotes: Precio vs. Ubicacion del motor')
plt.xlabel('Ubicación del motor')
plt.ylabel('Precio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
